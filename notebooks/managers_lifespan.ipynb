```python
# Rename the first column to 'Name' for all dataframes
arsenal.rename(columns={'Name/Date of birth': 'Name'}, inplace=True)
city.rename(columns={'Name/Date of birth': 'Name'}, inplace=True)
chelsea.rename(columns={'Name/Date of birth': 'Name'}, inplace=True)
united.rename(columns={'Name/Date of birth': 'Name'}, inplace=True)
```

```python
# Get current managers (index 0)
pep_tenure = city.iloc[0]['Time in post']
arteta_tenure = arsenal.iloc[0]['Time in post']

print(f"Pep: {pep_tenure}")
print(f"Arteta: {arteta_tenure}")
print("\n")

# Chelsea recent (top 6)
print("CHELSEA RECENT MANAGERS:")
for i in range(6):
    print(f"{chelsea.iloc[i]['Name']}: {chelsea.iloc[i]['Time in post']}")
print("\n")

# Man United post-2013 (let's see first 10)
print("MAN UTD POST-2013:")
for i in range(10):
    print(f"{united.iloc[i]['Name']}: {united.iloc[i]['Time in post']}")
```

```python
# For Chelsea - exclude very short tenures (under 1 month = interim) and Rosenior becasue he just got in
# Skip 1 (McFarlane), 4 (Lampard), 5 (Saltor)
chelsea_permanent = [2, 3]  # Rosenior, Maresca, Pochettino

print("CHELSEA PERMANENT:")
for i in chelsea_permanent:
    print(f"{chelsea.iloc[i]['Name']}: {chelsea.iloc[i]['Time in post']}")
print("\n")

# For Man United - exclude very short tenures
# 0 (Fletcher), 2 (van Nistelrooy), 4 (Rangnick), 5 (Carrick)
united_permanent = [1, 3, 6, 8, 9]  # Amorim, Ten Hag, Solskjaer, Mourinho, Van Gaal

print("MAN UTD PERMANENT:")
for i in united_permanent:
    print(f"{united.iloc[i]['Name']}: {united.iloc[i]['Time in post']}")
```
